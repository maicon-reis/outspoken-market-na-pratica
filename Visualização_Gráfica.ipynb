##**Primeiros Passos**

Ao longo deste código iremos percorrer as principais configurações da biblioteca Matplotlib, com isso nós poderemos criar gráficos com recursos que vão do básico ao avançado.

Começaremos importando a biblioteca que será utilizada para criação de gráficos, a Matplotlib. Para tornar mais prática a utilização realizaremos a importação do módulo pyplot e apelidaremos como ```plt``` e com isso simplificaremos o código.


```
# importando a biblioteca Matplotlib
import matplotlib.pyplot as plt
```

Além da biblioteca de criação de gráficos, importaremos as bibliotecas pandas e numpy, para tabulação de dados e realização de cálculos, respectivamente. Além de uma integração entre o Pandas e o Matplotlib para evitar incompatibilidades.

````python
# importando as bibliotecas
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline

# Adicionando integração entre pandas e matplotlib
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
````

Começaremos gerando um array contendo 50 números com espaçamento uniforme entre 0 e 5 com a função numpy ```linspace()```.

Em seguida criaremos mais dois arrays contendo o cosseno e o arco tangente dos dados anteriormente criados. Com isso já temos os dados para geração de gráficos. Poderíamos simplesmente importar um banco de dados, sim, mas fazendo desta forma exercitamos outras habilidades necessárias.

````python
# gerando dados a partir de um vetor
dados = np.linspace(0, 5, 50)

# cosseno do vetor dados
cos = np.cos(dados)

# arco tangente do vetor
arctan = np.arctan(dados)

# visualizando os dados criados
print('cos:\n{}\n\narctan:\n{}'.format(cos, arctan))
````

No primeiro gráfico criado, a versão mais simples criaremos um gráfico com dois plots na mesma figura utilizando a função matplotlib ```figure()```.

Denter vários, um dos principais parâmetros desta função especifica a largura e altura da figura em polegadas.

```
# sintaxe padrão
plt.figure(figsize=[6.4, 4.8])
```

O tipo de gráfico, também, será o mais simples, sem configurações complexas, contendo somente as coordenadas dos pontos (x e y) e o parâmetro opcional *fmt*, que nada mais é do que a utilização de atalhos para definir características básicas: marcador, estilo de linha e cor.

```
# Criando um gráfico - configuração mais básica
plt.plot([x], y, '[marcador][linha][cor]')
```

Entenda que a forma de especificação acima é uma mera conveniência, veremos a mesma plotagem poderia ter sido expressada de uma forma mais explícita.

```
# Criando um gráfico - exemplificação de forma extensa
plt.plot([x], y, marker=[marcador], linestyle=[linha], color=[cor])
```

Para ter uma lista com os tipos de marcadores linhas e cores são suportados, verificar a documentação do matplotlib.


Obs.: Para evitar que seja exibido o objeto da figura na tela, juntamente com o gráfico, podemos optar por incluir ponto-e-vírgula (;) na última linha de código, ou com a sintaxe ```plt.show()```.


````python
# criando um gráfico simples
fig = plt.figure(figsize=[10.5, 4.3]) # permite dois ou mais plots na mesma figura
plt.plot(dados, cos, "-.b") # criará x e y usando ponto e traco na cor azul
plt.plot(dados, arctan, "ko:"); # criará x e y usando o tracejado na cor preta
````

Como falei anteriormente o parâmetro ```fmt``` é opcional, mas existem muito outras configurações possíveis no plot, porém vou listar as mais utilizadas:
* ```alpha``` (escalar ou None) - O alpha deve ser estar dentro do intervalo entre 0 e 1 e corresponde a opacidade;
* ```color ou c``` - Configura a cor da linha;
* ```label``` (string) - Corresponde ao rótulo de dados que será exibido;
* ```linestyle ou ls``` - {'-' (sólido), '--' (tracejado), '-.' (ponto e vírgula), ':' (pontilhado), etc} Configura o estilo da linha;
* ```linewidth ou lw``` - (float) Configura a largura da linha;
* ```marker``` {'.' (ponto), ',' (pixel), 'o' (círculo), 'v' (triângulo para baixo), '^' (triângulo para cima), '*' (estrela), '|' (linha vertical), '_' (linha horizontal), etc } - Configura um marcador para a linha;
* ```markersize ou ms``` (float) - Configura o tamanho do marcador nos pontos.


````python
# criando um gráfico simples
fig = plt.figure(figsize=[10.5, 4.3]) # permite dois ou mais plots na mesma figura
plt.plot(dados, cos, ls='-', marker='.', c='b') # criará x e y usando ponto e traco na cor azul
plt.plot(dados, arctan, ls='--', marker='o', c='k'); # criará x e y usando o pontilhado na cor preta
````

### **Subplotagens**

Agora criarei dois gráficos verticalmente posicionados com o módulo ```plt.subplot()```.

````python
# duas séries em plots separadados
plt.figure()
plt.subplot(2, 1, 1) # linhas, colunas e painel
plt.plot(dados, arctan, "--", c="red")

plt.subplot(2, 1, 2)
plt.plot(dados, cos, "-");
````

Uma outra form de criar dois ou mais subplots pode-se também utilizar ```add_subplot()```, sendo que recebe três parâmetros principais, o primeiro representando o número de linhas do grid, o segundo o número de colunas, e o terceiro a posição do gráfico, da esquerda para direita de cima para baixo.

````python
# duas séries em plots separadados
fig = plt.figure()
ax1 = fig.add_subplot(2, 1, 1) # linhas, colunas e painel
ax1.plot(dados, arctan, "--", c="red")

ax2 = fig.add_subplot(2, 1, 2)
ax2.plot(dados, cos, "-");
````

Uma terceira forma de criar uma grade de subplotagens é utilizando o método ```plt.subplots()```. Esse método cria uma nova figura e devolve um array NumPy contendo os objetos de subplotage, criados com base nos parâmetros 1 e 2.

```
# Sintaxe básica com parâmetros padrão
plt.subplots(nrows=1, ncols=1, sharex=None, sharey=None)
```

Caso se deseje informar que as subplotagens compartilhes dos mesmos eixos podemos utilizar os parâmetros nomeados sharex e sharey, que podem receber os valores 'None', 'all', 'row', 'col'. Muito útil em séries temporais.

O espaçamento pode ser alterado usando o método ```subplot_adjust()``` dos objetos Figure. Entre os parâmetros recebidos temos o ```wspace``` e ```hspace``` que controlam o percentual de largura e altura da figura,respectivamente. 

````python
# duas séries em plots separadados
fig, axes = plt.subplots(2, 1, sharex='col', sharey='col')

axes[0].plot(dados, arctan, "--", c="red")
axes[1].plot(dados, cos, "-")

plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=None,
                    hspace=0) # Configurando o hspace=0 para entre as figuras
````

````python
fig, ax = plt.subplots(nrows=2, sharex='col', sharey='col')

ax[0].plot(arctan, color="red") # pode-se usar c ao invés de color
ax[1].plot(cos, color="green")

ax[0].set(xlim=(0, 50), ylim=(-2, 2), xlabel="x", ylabel="valores",
          title="Arco Tangente");
ax[1].set(xlim=(0, 50), ylim=(-2, 2), xlabel="y", ylabel="valores", title="Cosseno")

ax[0].legend(["Vetor Arctan"])
ax[1].legend(["Vetor Coss"])

fig.tight_layout() # afasta os gráficos e melhora a visualização
fig.show()
````

###**Ajustando os limites dos eixos**

Para ajustar os limites aos eixos x e y utilizamos as funções ```xlim()``` e ```ylim()``` dos eixos correntes.

```
# sintaxe padrão
xlim(botom, top)
ylim(bottom, top)
```
Nos casos de grades com subplots, para ajustar limites sem argumentos utilizamos a função ```get_ylim()```, já chamar a função com argumentos utiliza-se ```set_ylim()```.

````python
plt.plot(dados, label='Vetor A')
plt.plot(dados - 2, label="Vetor A deslocado")
plt.xlim(0, 50)
plt.ylim(-5, 8)
plt.legend()
plt.title("Gráfico de Vetores");
````

##**Histograma**

O histograma é um tipo de gráfico utilizado para visualizra distribuição. O módulo maplotlib utilizado para criar gráficos de histograma é o ```plt.hist()```.

```
# sintaxe básica
plt.hist(x, bins=None, color=None, label=None)
```

Principais parâmetros:
* ```x``` (array ou sequência de arrays) - Entrada de valores;
* ```bins``` (inteiro, sequência ou string) - Quantidade de barras nas quais os dados serão distribuídos. Caso seja uma string recebe entradas suportadas por ```numpy.histogram_bin_edges```;
* ```density``` (booleano) - Se configurado como ```True``` retorna uma densidade de probabilide;
* ```histtype``` {'bar', 'barstacked', 'step', 'stepfilled'}, default: 'bar' - Define o tipo de desenho do histograma;
* ```aling``` {'left', 'mid', 'right'} default: 'mid' - Define o alinhamento horizontal das barras;
* ```orientation``` {'vertical', 'horizontal'} default: 'vertical' -  Define a orientação do gráfico;
* ```color``` (cor, array de cores ou None) - Cor ou sequência de cores para ser utilizado no gráfico;
* ```label``` (String ou None) - Rótulo do gráfico;
* ```stacked``` (booleano) - Se configurado como True, múltiplas barras são empilhadas uma ao lado da outra.


````python
rand = np.random.randn(1000)

plt.hist(rand, bins=30, density=True, alpha=0.5, histtype="stepfilled",
         color="darkgreen", edgecolor='red', label="Gráfico de Histograma")

plt.legend()
plt.title("Gráfico de Histograma")
plt.show()
````

É possível criar configurações padrão para vários gráfico e assim não precisar ficar repetindo as mesmas configurações. Para tanto é necessário criar um dicionário contendo na chave o parâmetro nomeado e no valor o respectivo valor do parâmetro. Assim que realizar tais configurações chamá-lo dentro do plot precedido por dois asteríscos (kwargs).

````python
dist_1 = np.random.normal(4, 0.75, 500)
dist_2 = np.random.normal(-3, 1, 500)
dist_3 = np.random.normal(0.5, 2, 500)
dist_4 = np.random.normal(0.5, 0.5, 500)

# criando um dicionário de configurações padrão
config = {"histtype": "stepfilled", "alpha": 0.7, "density": True, "bins": 50}

plt.hist(dist_1, **config, label="Dist_1")
plt.hist(dist_2, **config, label="Dist_2")
plt.hist(dist_3, **config, label="Dist_3")
plt.hist(dist_4, **config, label="Dist_4")

# Formatação da legenda
plt.legend(loc="upper left", frameon=True, ncol=2, fancybox=True, framealpha=0.7,
           shadow=True, borderpad=0.5)
plt.title("Diversas Distribuições")
plt.show()
````

````python
import yfinance as yf
import datetime as dt

# carregando a base de dados
bvsp =  yf.download( "^BVSP", start=dt.datetime( 2010, 1, 1 ), end=dt.datetime.now() )
bvsp.shape
````

````python
with plt.style.context("seaborn-deep"):
    fig, ax = plt.subplots(nrows=1, figsize=(12, 5))
    ax.plot(bvsp.index, bvsp["Adj Close"], label="Ibovespa", linewidth=0.5)

    # Configurações de rótulos
    eixos_config = {'fontsize': 14, 'color': 'blue'}
    ax.set_xlabel("2010 a 2021", **eixos_config)
    ax.set_ylabel("Pontuação", **eixos_config)
    ax.legend(loc="lower right", frameon=True, ncol=2, fancybox=True, framealpha=0.7,
           shadow=True, borderpad=0.5)
    titulo_config = {'fontsize': 16, 'fontweight': 'bold', 'color': 'blue'}
    ax.set_title("Diversas Distribuições", **titulo_config)

    # Tornando as colunas superior e direita invisíveis
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

    # Definindo o limite esquerdo do eixo x
    ax.set_xlim(left=bvsp.index[0])

    # remover os xticks
    plt.tick_params(axis='both',
                    which='both',
                    bottom=False,
                    left=False,
                    labelbottom=True)
    plt.show()
````

````python
with plt.style.context("dark_background"):
    fig, ax = plt.subplots(nrows=1, figsize=(12, 5))
    ax.plot(bvsp.index, bvsp["Adj Close"], label="Ibovespa", linewidth=0.5)
    plt.xlabel("2010 a 2021")
    plt.ylabel("Pontuação")
    plt.legend(loc="lower right", frameon=True, ncol=2, fancybox=True, framealpha=0.7,
           shadow=True, borderpad=0.5)
    plt.title("Diversas Distribuições")
    plt.show()
````

````python
with plt.style.context("dark_background"):
    plt.figure(figsize=(12, 5))
    plt.plot(bvsp.index, bvsp["Adj Close"], label="Ibovespa", linewidth=1)
    plt.xlabel("2010 a 2021")
    plt.ylabel("Pontuação")
    plt.legend(loc="lower right", frameon=True, ncol=2, fancybox=True, framealpha=0.7,
           shadow=True, borderpad=0.5)
    plt.title("Gráfico com anotações")

setup  = dict(size=8, color='yellow', fontsize=12, fontweight='bold')
setup2 = dict(color='yellow', linestyle="--", linewidth=0.5)

plt.text(y, x, x, **setup)
plt.text("2010-01-01", 75000, "Início da era Dilma", **setup)
plt.axvline("2010-01-01", **setup2)
plt.text("2016-08-31", 75000, "Impeachment", **setup)
plt.axvline("2016-08-31", **setup2)
plt.savefig("grafico.png")
````
